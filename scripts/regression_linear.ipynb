## Import Libs

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Regression

```python
# Create fake data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Converting to one-dimensional vectors
X = X.flatten()
y = y.flatten()

# Calculando as médias de X e y
X_mean = np.mean(X)
y_mean = np.mean(y)

# Calculating the slope (b1)
numerador = np.sum((X - X_mean) * (y - y_mean))
denominador = np.sum((X - X_mean) ** 2)
b1 = numerador / denominador

#Calculating the intercept (b0)
b0 = y_mean - b1 * X_mean

# Displaying the coefficients
print(f"Intercepto (b0): {b0:.2f}")
print(f"Coeficiente angular (b1): {b1:.2f}")

# Making predictions
y_pred = b0 + b1 * X

# Viewing the results
plt.scatter(X, y, color="blue", label="Real data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Line of Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Simple Linear Regression without sklearn")
plt.show()
```

# Calculation of error metrics for the model

```python
# Calculating assessment metrics
n = len(y)

# Mean Absolute Error (MAE)
mae = np.mean(np.abs(y - y_pred))

# Mean Square Error 
mse = np.mean((y - y_pred) ** 2)

# Root Mean Square Error
rmse = np.sqrt(mse)

# Coefficient of Determination (R²)
ss_total = np.sum((y - y_mean) ** 2)  # Sum of total squares
ss_residual = np.sum((y - y_pred) ** 2)  # Sum of squares of residuals
r2 = 1 - (ss_residual / ss_total)

# View results

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Square Error  (MSE): {mse:.2f}")
print(f"Root Mean Square Error: {rmse:.2f}")
print(f"Coefficient of Determination (R²): {r2:.4f}")
```

# Sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Generating synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)


# Scikit-learn Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Extracting coefficients
b0_sklearn = model.intercept_[0]
b1_sklearn = model.coef_[0][0]

# Making predictions
y_pred_sklearn = model.predict(X)

# Computing evaluation metrics
mae_sklearn = mean_absolute_error(y, y_pred_sklearn)
mse_sklearn = mean_squared_error(y, y_pred_sklearn)
rmse_sklearn = np.sqrt(mse_sklearn)
r2_sklearn = r2_score(y, y_pred_sklearn)


print("Scikit-learn Linear Regression:")
print(f"Intercept (b0): {b0_sklearn:.2f}")
print(f"Slope (b1): {b1_sklearn:.2f}")
print(f"Mean Absolute Error (MAE): {mae_sklearn:.2f}")
print(f"Mean Squared Error (MSE): {mse_sklearn:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_sklearn:.2f}")
print(f"R-squared (R²): {r2_sklearn:.4f}\n")

# Plotting results
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, y_pred_sklearn, color="green", linestyle="dashed", linewidth=2, label="Sklearn Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Scikit-learn Linear Regression")
plt.show()
```

## Import Libs

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Regression

```python
# Create fake data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Converting to one-dimensional vectors
X = X.flatten()
y = y.flatten()

# Calculando as médias de X e y
X_mean = np.mean(X)
y_mean = np.mean(y)

# Calculating the slope (b1)
numerador = np.sum((X - X_mean) * (y - y_mean))
denominador = np.sum((X - X_mean) ** 2)
b1 = numerador / denominador

#Calculating the intercept (b0)
b0 = y_mean - b1 * X_mean

# Displaying the coefficients
print(f"Intercepto (b0): {b0:.2f}")
print(f"Coeficiente angular (b1): {b1:.2f}")

# Making predictions
y_pred = b0 + b1 * X

# Viewing the results
plt.scatter(X, y, color="blue", label="Real data")
plt.plot(X, y_pred, color="red", linewidth=2, label="Line of Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Simple Linear Regression without sklearn")
plt.show()
```

# Calculation of error metrics for the model

```python
# Calculating assessment metrics
n = len(y)

# Mean Absolute Error (MAE)
mae = np.mean(np.abs(y - y_pred))

# Mean Square Error 
mse = np.mean((y - y_pred) ** 2)

# Root Mean Square Error
rmse = np.sqrt(mse)

# Coefficient of Determination (R²)
ss_total = np.sum((y - y_mean) ** 2)  # Sum of total squares
ss_residual = np.sum((y - y_pred) ** 2)  # Sum of squares of residuals
r2 = 1 - (ss_residual / ss_total)

# View results

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Square Error  (MSE): {mse:.2f}")
print(f"Root Mean Square Error: {rmse:.2f}")
print(f"Coefficient of Determination (R²): {r2:.4f}")
```
